```python
Book_Store['Price']= Book_Store['Price'].astype(str)
```

```python
Book_Store['Price'] = Book_Store['Price'].fillna('').astype(str)
```

```python

Book_Store['Price'] = Book_Store['Price'].replace('<NA>', np.nan)
```

```python
# Remove comma from values in price column and change the type of float 64
print('Removing comma and change the data type to float')
Book_Store['Price'] = Book_Store['Price'].str.replace(',','').astype('float64')
Book_Store.dtypes
```

```python
Book_Store['Total Ratings'] = Book_Store['Total Ratings'].fillna('').astype(str)
```

```python
# remove comma from values in total ratings column and change the type int64
print('remove comma from total ratings and change data type to int64')
Book_Store['Total Ratings']=Book_Store['Total Ratings'].str.replace(',','').astype('int64')
Book_Store.dtypes
```

```python
# check how many missing values are there in all columns
Book_Store.isnull().sum()
```

```python
# replace missing value in language with the most occuring language
common_lang = Book_Store['Language'].mode()[0]
Book_Store['Language'].fillna(common_lang , inplace=True)
Book_Store.isnull().sum()
```

```python
# remove outliers (using IQR method)
# extra (above two questions with outliers removed)
def remove_outliers(df, column, multiplier=1.5):
# calculate q1 (25th percentile) and q3 (75th percentile) of the column
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)


# calculate IQR
    IQR = Q3 - Q1
 # define the lower and upper bounds for outliers
    lower_bound = Q1 - multiplier * IQR
    upper_bound = Q3 + multiplier * IQR
# Remove outliers by filtering the column with in the bounds
    df_cleaned = df[(df[column]>= lower_bound) & (df[column] <= upper_bound)]
    return df_cleaned
    
original_rows = len(Book_Store)
Book_Store = remove_outliers(Book_Store,'Price')
Book_Store = remove_outliers(Book_Store,'Pages')
rows_removed = original_rows - len(Book_Store)
print(f'number of rows removed: {rows_removed}')

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.boxplot(y=Book_Store['Price'])
plt.title('Boxplot of price (Outliers Removed)')
plt.subplot(1,2,2)
sns.boxplot(y=Book_Store['Pages'])
plt.title('Boxplot of pages(Outliers Removed)')
plt.tight_layout()
plt.show()
    
```
